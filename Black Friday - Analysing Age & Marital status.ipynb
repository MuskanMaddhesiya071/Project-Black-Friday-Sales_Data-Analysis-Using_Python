```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Purchase Distribution by Age')
```

### Conclusion

The bar chart displays the distribution of purchases across different age groups. The key observations are:

- The **26-35 age group** has the **highest number of purchases**, making it the most active segment.
- The **18-25** and **36-45** age groups also show substantial purchase activity, following the 26-35 group.
- Purchase activity gradually **declines with age** after 35, with significantly lower participation in the **46-50**, **51-55**, and **55+** age brackets.
- The **0-17** age group has the **lowest** number of purchases.

This suggests that the **young adult to middle-aged demographic (18-45)** is the primary driver of purchases, with peak activity in the 26-35 range.


```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
```

```python
data.plot.bar(x = 'Age', figsize = (8,6))
```

### Conclusion

The bar chart presents the count of unique products purchased across different age groups. Key insights include:

- The **26-35** age group purchased the **highest number of unique products**, indicating a strong diversity in their buying behavior.
- The **36-45**, **18-25**, and **46-50** age groups also show high product diversity, closely following the 26-35 segment.
- Even the **55+** and **51-55** groups have moderate product variety, suggesting interest in a range of products despite fewer total purchases.
- The **0-17** group has the **lowest number of unique products**, reflecting either limited buying capacity or a narrow interest range.

This data reinforces the significance of the **26-45 age range** as a key consumer demographic with both **high engagement** and **diverse product interest**.


```python
ax=df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (12, 6), title = 'Amount Spent by Age')
ax.set_xlabel('Age Group')
ax.set_ylabel('Total Purchase Amount (in Thousands)')
ax.set_yticklabels([f'{int(x/1000)}K' for x in ax.get_yticks()])
```

### Conclusion  

The bar chart titled **"Amount Spent by Age"** displays the total purchase amounts (in thousands) across different age groups. Key insights include:  

- The age group **26-35** spent the highest amount (**2,250,000K**), making them the top spenders.  
- The **18-25** group shows relatively lower spending, suggesting they may have less disposable income or different purchasing priorities.  
- The **36-45** age group represent a secondary high-spending segment.  
- Older groups (**46-50, 51-55, 55+**) exhibit declining expenditures, possibly due to reduced consumption or financial priorities shifting away from discretionary spending.  

**Recommendation:** Businesses should focus marketing efforts on **26-35**, while also exploring opportunities in the **36-45** demographic.

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'bar', figsize = (12, 6), title = ' Average Amount Spend by Age')
```


## Conclusion on Average Spending by Age  

- The chart illustrates a clear pattern in consumer spending across different age groups. Spending gradually increases with age until it peaks in the 51-55 age bracket, followed by a slight decrease in the 55+ category.
- The youngest group (0-17) shows the lowest average spending at approximately 8,900, 
- while the 51-55 age group demonstrates - the highest expenditure at roughly 9,600. This represents about a 7.9% increase from the youngest to the highest-spending age group.

- Interestingly, spending remains relatively stable across middle-age brackets (26-35, 36-45, and 46-50), with only minor variations between them. 

- The data suggests that consumer spending power generally strengthens with age until pre-retirement years, after which it moderately declines.
